```python
BASE_URL = 'https://habr.com'
```

```python
resp = requests.get(BASE_URL + '/ru/search/', params=params)
resp.status_code
```

```python
soup = BeautifulSoup(resp.text, 'lxml')
soup
```

```python
print(soup.prettify())
```

```python
article  = soup.find_all('article')
article
```

```python
type(article)
```

```python
results = []
for article_tag in article:
    datetime = article_tag.find('time').get('title')
    article_name = article_tag.find('h2').get_text()
    # text_a = article.find('div', class_= 'tm-article-body')

    link = BASE_URL + article_tag.find('a').get('href')
    results.append([datetime, article_name, link])
    # print(datetime)
    # print(article_name)
    # print(text_a)
    # print(link)
    # print('*'*100)
results
```

```python
Base = pd.DataFrame(results, columns =['Date', 'Title', 'link'])
Base
```
